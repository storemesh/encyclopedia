````python
def create_article(consolidated_responses):
    article = ""
    for topic, keywords in consolidated_responses.items():
        article += f"# {topic}\n"
        for keyword, details in keywords.items():
            has_content = False
            article += f"## {keyword}\n"

            introduction = details.get('Introduction')
            if introduction and introduction != 'N/A':
                article += f"### Introduction\n\t{introduction}\n\n"
                has_content = True

            origin_discovery = details.get('Origin and Discovery')
            if origin_discovery and origin_discovery != 'N/A':
                article += f"### Origin and Discovery\n\t{origin_discovery}\n\n"
                has_content = True

            properties = details.get('Details', {}).get('Properties')
            types_variants = details.get('Details', {}).get('Types and Variants')
            if properties or types_variants:
                article += f"### Properties\n"
                if properties and properties != 'N/A':
                    article += f"\t- Properties: {properties}\n"
                    has_content = True
                if types_variants and types_variants != 'N/A':
                    article += f"\t- Types and Variants: {types_variants}\n"
                    has_content = True
                article += '\n'

            applications = details.get('Applications and Use Cases')
            if applications and applications != 'N/A':
                article += f"### Applications and Use Cases\n\t{applications}\n\n"
                has_content = True

            care_maintenance = details.get('Care and Maintenance', details.get('Further Information'))
            if care_maintenance and care_maintenance != 'N/A':
                article += f"### Care and Maintenance or Further Information\n\t{care_maintenance}\n\n"
                has_content = True

            analysis_examination = details.get('Analysis and Examination')
            if analysis_examination and analysis_examination != 'N/A':
                article += f"### Analysis and Examination\n\t{analysis_examination}\n\n"
                has_content = True

            popularity_influence = details.get('Popularity and Influence')
            if popularity_influence and popularity_influence != 'N/A':
                article += f"### Popularity and Influence\n\t{popularity_influence}\n\n"
                has_content = True

            conclusion = details.get('Conclusion')
            if conclusion and conclusion != 'N/A':
                article += f"### Conclusion\n\t{conclusion}\n\n"
                has_content = True
            if has_content:
                article += "\n"
    
    return article
article = create_article(consolidated_responses)
print(article)
````

````python
from docx import Document
from docx.shared import Pt

def save_all_articles_to_docx(consolidated_responses, filename):
    doc = Document()
    
    for topic, keywords in consolidated_responses.items():
        # Convert JSON response to article text
        article = create_article(consolidated_responses)
        
        # Split article into sections by the topic headers
        sections = article.split('\n')
        
        for section in sections:
            if section.startswith('# '):  # Topic header
                header = section[2:].strip()
                para = doc.add_paragraph()
                run = para.add_run(header)
                para.style = 'Heading 1'
                run.font.size = Pt(18)
                run.bold = True
            elif section.startswith('## '):  # Subtopic header
                header = section[3:].strip()
                para = doc.add_paragraph()
                run = para.add_run(header)
                para.style = 'Heading 2'
                run.font.size = Pt(18)
                run.bold = True
            elif section.startswith('### '):  # Sub-subtopic header
                subheader = section[4:].strip()
                para = doc.add_paragraph()
                run = para.add_run(subheader)
                para.style = 'Heading 3'
                run.font.size = Pt(16)
                run.bold = True
            else:  # Content
                para = doc.add_paragraph(section)
                for run in para.runs:
                    run.font.size = Pt(14)
    
    doc.save(filename)

# Example usage
filename = 'all_articles_4.docx'
save_all_articles_to_docx(consolidated_responses, filename)
print(f"Saved all articles to {filename}")
````

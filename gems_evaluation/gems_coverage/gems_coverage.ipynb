```python
from fuzzywuzzy import fuzz
```

```python
path = r'C:\Users\DELL\Desktop\gems_evaluation'
file_gems_encyclopedia = f'{path}\Gems_Encyclopedia.csv'
file_gems_tfidf = f'{path}\Gems_TFIDF.csv'
file_gems_lda = f'{path}\Gems_LDA.csv'
file_gems_llm = f'{path}\Gems_LLM.csv'

df_encyclopedia = pd.read_csv(file_gems_encyclopedia)
df_tfidf = pd.read_csv(file_gems_tfidf)
df_lda = pd.read_csv(file_gems_lda)
df_llm = pd.read_csv(file_gems_llm)
```

```python
keywords_encyclopedia = set(df_encyclopedia['keyword'])
keywords_tfidf = set(df_tfidf['keyword'])
keywords_lda = set(df_lda['keyword'])
keywords_llm = set(df_llm['keyword'])
```

```python
def is_similar(word, keywords_set, threshold=90):
    for keyword in keywords_set:
        if fuzz.ratio(word, keyword) >= threshold:
            return 1
    return 0
```

```python
results = pd.DataFrame(df_encyclopedia['keyword'])
results['Gems_TFIDF'] = results['keyword'].apply(lambda x: is_similar(x, keywords_tfidf))
results['Gems_LDA'] = results['keyword'].apply(lambda x: is_similar(x, keywords_lda))
results['Gems_LLM'] = results['keyword'].apply(lambda x: is_similar(x, keywords_llm))
```

```python
results
```

```python
total_keywords = len(df_encyclopedia)

count_tfidf = results['Gems_TFIDF'].sum()
count_lda = results['Gems_LDA'].sum()
count_llm = results['Gems_LLM'].sum()
count_tfidf, count_lda, count_llm
```

```python
coverage_tfidf = (count_tfidf / total_keywords) * 100
coverage_lda = (count_lda / total_keywords) * 100
coverage_llm = (count_llm / total_keywords) * 100
coverage_tfidf, coverage_lda, coverage_llm
```

```python
words_with_1_tfidf = results[results['Gems_TFIDF'] == 1]['keyword']
words_with_1_lda = results[results['Gems_LDA'] == 1]['keyword']
words_with_1_llm = results[results['Gems_LLM'] == 1]['keyword']

print("Words with 1 in Gems_TFIDF:")
print(words_with_1_tfidf.tolist())

print("\nWords with 1 in Gems_LDA:")
print(words_with_1_lda.tolist())

print("\nWords with 1 in Gems_LLM:")
print(words_with_1_llm.tolist())
```

```python
from fuzzywuzzy import fuzz
```

```python
path = r'C:\Users\kai_k\Documents\intern_datastoremesh\jewelry'
file_gems_encyclopedia = 'Encyclopedia_Jewelry.csv'
file_gems_ontology = 'Ontology.csv'

df_encyclopedia = pd.read_csv(file_gems_encyclopedia)
df_ontology = pd.read_csv(file_gems_ontology)
```

```python
print(df_encyclopedia.columns)
print(df_ontology.columns)
```

```python
keywords_encyclopedia = set(df_encyclopedia['Attributes'])
keywords_ontology = set(df_ontology['Attributes'])
```

```python
def is_similar(word, keywords_set, threshold=90):
    for keyword in keywords_set:
        if fuzz.ratio(word, keyword) >= threshold:
            return 1
    return 0
```

```python
results = pd.DataFrame(df_encyclopedia['Attributes'])
results['Gems_ontology'] = results['Attributes'].apply(lambda x: is_similar(x, keywords_ontology))
```

```python
results
```

```python
total_keywords = len(df_encyclopedia)

count_ontology = results['Gems_ontology'].sum()

count_ontology
```

```python
coverage_ontology = (count_ontology / total_keywords) * 100

coverage_ontology
```
